<div style="margin-top: 40px; background-color: #f0f4f8; padding: 30px; border-radius: 15px; border: 1px solid #d1d9e6;"> <h3 style="text-align: center; color: #2c3e50; margin-bottom: 20px;">‚öñÔ∏è ‡∏ó‡∏≥‡πÑ‡∏°‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏Å Semi-supervised?</h3> <div style="display: flex; justify-content: space-around; flex-wrap: wrap; text-align: center; gap: 20px;"> <div style="flex: 1; min-width: 200px;"> <strong style="color: #2980b9; font-size: 1.2em;">Cost Efficiency</strong> <p style="font-size: 0.9em; color: #666;">‡∏õ‡∏£‡∏∞‡∏´‡∏¢‡∏±‡∏î‡∏Ñ‡πà‡∏≤‡πÅ‡∏£‡∏á‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≥ Labeling ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡∏ú‡∏π‡πâ‡πÄ‡∏ä‡∏µ‡πà‡∏¢‡∏ß‡∏ä‡∏≤‡∏ç‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ó‡∏≤‡∏á</p> </div> <div style="flex: 1; min-width: 200px;"> <strong style="color: #27ae60; font-size: 1.2em;">Better Generalization</strong> <p style="font-size: 0.9em; color: #666;">‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏´‡πá‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢‡∏Å‡∏ß‡πà‡∏≤ ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡πÑ‡∏î‡πâ‡∏î‡∏µ‡∏Ç‡∏∂‡πâ‡∏ô‡πÉ‡∏ô‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏à‡∏£‡∏¥‡∏á</p> </div> <div style="flex: 1; min-width: 200px;"> <strong style="color: #e67e22; font-size: 1.2em;">Leverage Big Data</strong> <p style="font-size: 0.9em; color: #666;">‡πÉ‡∏ä‡πâ‡∏õ‡∏£‡∏∞‡πÇ‡∏¢‡∏ä‡∏ô‡πå‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏°‡∏´‡∏≤‡∏®‡∏≤‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ß‡πâ‡πÅ‡∏ï‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÄ‡∏ß‡∏•‡∏≤‡πÑ‡∏õ‡∏ï‡∏¥‡∏î‡∏õ‡πâ‡∏≤‡∏¢‡∏Å‡∏≥‡∏Å‡∏±‡∏ö</p> </div> </div> </div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.ensemble import RandomForestClassifier

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏°‡∏°‡∏ï‡∏¥ X (‡∏û‡∏¥‡∏Å‡∏±‡∏î) ‡πÅ‡∏•‡∏∞ y (‡πÄ‡∏â‡∏•‡∏¢)
np.random.seed(42)
X = np.random.rand(100, 2)  # ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 100 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ ‡∏°‡∏µ 2 ‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Label ‡∏à‡∏£‡∏¥‡∏á (‡∏Ñ‡∏£‡∏∂‡πà‡∏á‡πÅ‡∏£‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏° 0, ‡∏Ñ‡∏£‡∏∂‡πà‡∏á‡∏´‡∏•‡∏±‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏° 1)
y_true = np.array([0] * 50 + [1] * 50)

# 2. ‡πÅ‡∏Å‡∏•‡πâ‡∏á‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• "‡∏´‡∏≤‡∏¢‡πÑ‡∏õ" 80% (‡∏Å‡∏•‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô -1)
y_mixed = y_true.copy()
random_unlabeled_indices = np.random.choice(100, 80, replace=False)
y_mixed[random_unlabeled_indices] = -1

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡πÄ‡∏ó‡∏£‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏• Self-training
base_model = RandomForestClassifier()
self_training_model = SelfTrainingClassifier(base_model, threshold=0.75)
self_training_model.fit(X, y_mixed)

# 4. ‡∏î‡∏π‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå: ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏≤‡∏¢ Label ‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏≠‡∏∞‡πÑ‡∏£‡∏ö‡πâ‡∏≤‡∏á
y_final = self_training_model.transduction_  # ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ó‡∏µ‡πà‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏ï‡∏¥‡∏°‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏ö‡πÅ‡∏•‡πâ‡∏ß

# 5. ‡∏ß‡∏≤‡∏î‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö
plt.figure(figsize=(12, 5))

# ‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏£‡∏Å: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏≠‡∏ô‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô (‡∏°‡∏µ‡∏à‡∏∏‡∏î‡∏™‡∏µ‡πÄ‡∏ó‡∏≤‡∏Ñ‡∏∑‡∏≠ Unlabeled)
plt.subplot(1, 2, 1)
plt.scatter(X[y_mixed == 0, 0], X[y_mixed == 0, 1], c='blue', label='Labeled (0)')
plt.scatter(X[y_mixed == 1, 0], X[y_mixed == 1, 1], c='red', label='Labeled (1)')
plt.scatter(X[y_mixed == -1, 0], X[y_mixed == -1, 1], c='lightgray', label='Unlabeled')
plt.title("Before Self-training (Mostly Unlabeled)")
plt.legend()

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ä‡πà‡∏ß‡∏¢‡πÄ‡∏ï‡∏¥‡∏° Label ‡πÉ‡∏´‡πâ
plt.subplot(1, 2, 2)
plt.scatter(X[y_final == 0, 0], X[y_final == 0, 1], c='blue')
plt.scatter(X[y_final == 1, 0], X[y_final == 1, 1], c='red')
plt.title("After Self-training (All Labeled)")

plt.show()
```

<div style="margin-bottom: 30px; padding: 0 10px;"> <h3 style="color: #0369a1; border-bottom: 2px solid #0ea5e9; display: inline-block; padding-bottom: 5px;">üìç ‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á</h3> <p style="margin-top: 15px; color: #444; line-height: 1.7;"> ‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó E-commerce ‡∏°‡∏µ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ 100,000 ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° ‡πÅ‡∏ï‡πà‡∏°‡∏µ‡∏ó‡∏µ‡∏°‡∏á‡∏≤‡∏ô‡∏à‡πâ‡∏≤‡∏á‡∏°‡∏≤‡∏ï‡∏£‡∏ß‡∏à‡πÅ‡∏•‡∏∞‡∏ï‡∏¥‡∏î‡∏õ‡πâ‡∏≤‡∏¢‡∏Å‡∏≥‡∏Å‡∏±‡∏ö (Label) ‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô <b>"‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÄ‡∏ä‡∏¥‡∏á‡∏ö‡∏ß‡∏Å"</b> ‡∏´‡∏£‡∏∑‡∏≠ <b>"‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡πÄ‡∏ä‡∏¥‡∏á‡∏•‡∏ö"</b> ‡πÑ‡∏î‡πâ‡πÄ‡∏û‡∏µ‡∏¢‡∏á 1,000 ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡πÄ‡∏£‡∏≤‡∏à‡∏∂‡∏á‡πÉ‡∏ä‡πâ Self-training ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏• "‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏à‡∏≤‡∏Å‡∏á‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß" ‡∏°‡∏≤‡∏ä‡πà‡∏ß‡∏¢‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏á‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠: </p> <ul style="color: #555; line-height: 1.8;"> <li><b>Labeled Data (1%):</b> ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡∏ó‡∏µ‡πà‡∏Ñ‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡πÅ‡∏•‡πâ‡∏ß ‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏∏‡∏î‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏û‡∏≠‡∏à‡∏∞‡πÅ‡∏¢‡∏Å‡πÅ‡∏¢‡∏∞‡∏≠‡∏≠‡∏Å</li> <li><b>Unlabeled Data (99%):</b> ‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡∏°‡∏´‡∏≤‡∏®‡∏≤‡∏•‡∏ó‡∏µ‡πà‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÉ‡∏Ñ‡∏£‡∏≠‡πà‡∏≤‡∏ô ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏à‡∏∞‡∏Ñ‡πà‡∏≠‡∏¢‡πÜ ‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£</li> <li><b>Pseudo-Labeling:</b> ‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡πÇ‡∏°‡πÄ‡∏î‡∏• "‡πÄ‡∏î‡∏≤" ‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ä‡∏∑‡πà‡∏≠ ‡πÅ‡∏•‡πâ‡∏ß‡∏ô‡∏≥‡∏°‡∏≤‡∏™‡∏≠‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏≠‡∏á‡∏ï‡πà‡∏≠</li> </ul> </div>

<div style="background-color: #f0f9ff; border-left: 6px solid #0ea5e9; padding: 20px; border-radius: 8px;"> <h3 style="color: #0c4a6e; margin-top: 0;">üîç 3 ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á Self-training</h3> <div style="margin-bottom: 15px;"> <strong style="color: #0369a1;">1. Initial Training:</strong> <p style="margin: 5px 0; color: #444;">‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 1,000 ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏£‡∏Å‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Ñ‡∏ô‡πÄ‡∏â‡∏•‡∏¢‡πÑ‡∏ß‡πâ‡πÉ‡∏´‡πâ‡πÅ‡∏•‡πâ‡∏ß ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏é‡πÄ‡∏Å‡∏ì‡∏ë‡πå‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤</p> </div> <div style="margin-bottom: 15px;"> <strong style="color: #0284c7;">2. High-Confidence Prediction:</strong> <p style="margin: 5px 0; color: #444;">‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÑ‡∏õ‡πÑ‡∏•‡πà‡∏≠‡πà‡∏≤‡∏ô‡∏£‡∏µ‡∏ß‡∏¥‡∏ß‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏≠‡∏µ‡∏Å 99,000 ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏° ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏°‡∏±‡∏ô <b>"‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡∏°‡∏≤‡∏Å" (‡πÄ‡∏ä‡πà‡∏ô ‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡πÄ‡∏Å‡∏¥‡∏ô 95%)</b> ‡∏°‡∏≤‡πÉ‡∏™‡πà‡∏õ‡πâ‡∏≤‡∏¢‡∏Å‡∏≥‡∏Å‡∏±‡∏ö‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏á (Pseudo-label)</p> </div> <div> <strong style="color: #0ea5e9;">3. Re-training Loop:</strong> <p style="margin: 5px 0; color: #444;">‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ô‡∏≥‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏±‡∏ô‡πÄ‡∏û‡∏¥‡πà‡∏á‡πÄ‡∏î‡∏≤‡πÉ‡∏´‡πâ‡∏ï‡∏±‡∏ß‡πÄ‡∏≠‡∏á‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏Å‡∏µ‡πâ ‡∏Å‡∏•‡∏±‡∏ö‡∏°‡∏≤‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡πÅ‡∏£‡∏Å ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ó‡∏£‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏≠‡∏á‡πÉ‡∏´‡∏°‡πà‡πÉ‡∏´‡πâ‡∏â‡∏•‡∏≤‡∏î‡∏Ç‡∏∂‡πâ‡∏ô ‡∏ó‡∏≥‡∏ã‡πâ‡∏≥‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢‡πÜ ‡∏à‡∏ô‡∏Ñ‡∏£‡∏ö‡∏ó‡∏∏‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•</p> </div> </div> </div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏°‡∏°‡∏ï‡∏¥ 100 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£
np.random.seed(42)
n_samples = 100

# View 1: ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡πÉ‡∏ô‡πÄ‡∏ß‡πá‡∏ö (‡πÄ‡∏ä‡πà‡∏ô ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≥‡∏ß‡∏¥‡∏ä‡∏≤‡∏Å‡∏≤‡∏£)
X1 = np.random.randn(n_samples, 2) + np.array([0, 0]) # ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏ö‡∏±‡∏ô‡πÄ‡∏ó‡∏¥‡∏á
X1[50:] += 3 # ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏®‡∏∂‡∏Å‡∏©‡∏≤

# View 2: ‡∏Ñ‡∏≥‡∏ö‡∏ô‡∏•‡∏¥‡∏á‡∏Å‡πå (‡πÄ‡∏ä‡πà‡∏ô ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏ô‡∏Å‡∏î‡πÅ‡∏ä‡∏£‡πå/‡∏•‡∏¥‡∏á‡∏Å‡πå‡∏≠‡πâ‡∏≤‡∏á‡∏≠‡∏¥‡∏á)
X2 = np.random.randn(n_samples, 2) + np.array([0, 0])
X2[50:] += 3

# Label ‡∏à‡∏£‡∏¥‡∏á: 0 = ‡∏ö‡∏±‡∏ô‡πÄ‡∏ó‡∏¥‡∏á, 1 = ‡∏Å‡∏≤‡∏£‡∏®‡∏∂‡∏Å‡∏©‡∏≤
y_true = np.array([0]*50 + [1]*50)

# 2. ‡πÅ‡∏Å‡∏•‡πâ‡∏á‡∏ó‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡πà‡∏≤‡πÄ‡∏£‡∏≤‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢‡∏Å‡∏≥‡∏Å‡∏±‡∏ö‡πÅ‡∏Ñ‡πà 6 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ (‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏•‡∏∞ 3) ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏õ‡πá‡∏ô -1 (Unlabeled)
y_train = np.full(n_samples, -1)
y_train[:3] = 0
y_train[50:53] = 1

# 3. ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏£‡∏∞‡∏ö‡∏ß‡∏ô‡∏Å‡∏≤‡∏£ Co-training (Loop 10 ‡∏£‡∏≠‡∏ö)
clf1 = GaussianNB() # ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö View 1
clf2 = GaussianNB() # ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö View 2

for i in range(10):
    # ‡∏Ñ‡∏±‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢‡∏Å‡∏≥‡∏Å‡∏±‡∏ö‡πÉ‡∏ô‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô
    labeled_idx = np.where(y_train != -1)[0]
    unlabeled_idx = np.where(y_train == -1)[0]
    
    if len(unlabeled_idx) == 0: break
    
    # ‡πÄ‡∏ó‡∏£‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏î‡πâ‡∏ß‡∏¢‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢
    clf1.fit(X1[labeled_idx], y_train[labeled_idx])
    clf2.fit(X2[labeled_idx], y_train[labeled_idx])
    
    # ‡πÉ‡∏´‡πâ Clf1 ‡∏ó‡∏≤‡∏¢‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢ ‡πÅ‡∏•‡∏∞‡∏´‡∏≤‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà "‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î"
    probs1 = clf1.predict_proba(X1[unlabeled_idx])
    max_probs1 = np.max(probs1, axis=1)
    best_idx1 = unlabeled_idx[np.argmax(max_probs1)]
    
    # ‡πÉ‡∏´‡πâ Clf2 ‡∏ó‡∏≤‡∏¢‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢ ‡πÅ‡∏•‡∏∞‡∏´‡∏≤‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà "‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î"
    probs2 = clf2.predict_proba(X2[unlabeled_idx])
    max_probs2 = np.max(probs2, axis=1)
    best_idx2 = unlabeled_idx[np.argmax(max_probs2)]
    
    # "‡πÅ‡∏•‡∏Å‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ñ‡∏ß‡∏≤‡∏°‡∏£‡∏π‡πâ": ‡πÄ‡∏≠‡∏≤‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà Clf1 ‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à ‡πÑ‡∏õ‡πÉ‡∏™‡πà‡∏õ‡πâ‡∏≤‡∏¢‡πÉ‡∏´‡πâ Clf2 ‡πÅ‡∏•‡∏∞‡πÉ‡∏ô‡∏ó‡∏≤‡∏á‡∏Å‡∏•‡∏±‡∏ö‡∏Å‡∏±‡∏ô
    y_train[best_idx1] = clf1.predict(X1[best_idx1].reshape(1, -1))[0]
    y_train[best_idx2] = clf2.predict(X2[best_idx2].reshape(1, -1))[0]

# 4. ‡∏ß‡∏±‡∏î‡∏ú‡∏•‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏°‡πà‡∏ô‡∏¢‡∏≥‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
y_pred = clf1.predict(X1)
print(f"Accuracy after Co-training: {accuracy_score(y_true, y_pred) * 100}%")

# 5. ‡∏ß‡∏≤‡∏î‡∏Å‡∏£‡∏≤‡∏ü‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
plt.figure(figsize=(10, 5))
plt.scatter(X1[y_pred==0, 0], X1[y_pred==0, 1], c='red', label='Predict: Entertainment', alpha=0.5)
plt.scatter(X1[y_pred==1, 0], X1[y_pred==1, 1], c='blue', label='Predict: Education', alpha=0.5)
plt.title("Co-training Result: Classifier A learned from Classifier B's confidence")
plt.legend()
plt.show()
```

<div style="margin-bottom: 30px; padding: 0 10px;"> <h3 style="color: #6b21a8; border-bottom: 2px solid #a855f7; display: inline-block; padding-bottom: 5px;">üìç ‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á</h3> <p style="margin-top: 15px; color: #444; line-height: 1.7;"> ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÅ‡∏¢‡∏Å‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏ã‡∏ï‡πå‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô <b>"‡πÄ‡∏ß‡πá‡∏ö‡∏Å‡∏≤‡∏£‡∏®‡∏∂‡∏Å‡∏©‡∏≤"</b> ‡∏´‡∏£‡∏∑‡∏≠ <b>"‡πÄ‡∏ß‡πá‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ö‡∏±‡∏ô‡πÄ‡∏ó‡∏¥‡∏á"</b> ‡πÇ‡∏î‡∏¢‡πÄ‡∏£‡∏≤‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 2 ‡∏°‡∏∏‡∏°‡∏°‡∏≠‡∏á (Views) ‡∏ó‡∏µ‡πà‡πÅ‡∏¢‡∏Å‡∏à‡∏≤‡∏Å‡∏Å‡∏±‡∏ô‡∏ä‡∏±‡∏î‡πÄ‡∏à‡∏ô: </p> <ul style="color: #555; line-height: 1.8;"> <li><b>View 1 (‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤):</b> ‡∏Ñ‡∏≥‡∏ï‡πà‡∏≤‡∏á‡πÜ ‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏≤‡∏Å‡∏è‡∏ö‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÄ‡∏ß‡πá‡∏ö‡∏ô‡∏±‡πâ‡∏ô‡πÜ</li> <li><b>View 2 (Hyperlink):</b> ‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏ß‡πá‡∏ö‡∏≠‡∏∑‡πà‡∏ô‡πÉ‡∏ä‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏î‡∏•‡∏¥‡∏á‡∏Å‡πå‡∏°‡∏≤‡∏´‡∏≤‡πÄ‡∏ß‡πá‡∏ö‡∏ô‡∏µ‡πâ (Anchor Text)</li> </ul> <p style="color: #444;">‡∏´‡∏±‡∏ß‡πÉ‡∏à‡∏Ñ‡∏∑‡∏≠: ‡∏ñ‡∏∂‡∏á‡πÅ‡∏°‡πâ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÑ‡∏°‡πà‡πÄ‡∏´‡πá‡∏ô‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡πÉ‡∏ô‡πÄ‡∏ß‡πá‡∏ö (View 1) ‡πÅ‡∏ï‡πà‡∏ñ‡πâ‡∏≤‡πÄ‡∏´‡πá‡∏ô‡∏Ñ‡∏≥‡∏ó‡∏µ‡πà‡∏Ñ‡∏ô‡∏≠‡∏∑‡πà‡∏ô‡∏•‡∏¥‡∏á‡∏Å‡πå‡∏°‡∏≤‡∏´‡∏≤ (View 2) ‡πÄ‡∏£‡∏≤‡∏Å‡πá‡∏û‡∏≠‡∏à‡∏∞‡πÄ‡∏î‡∏≤‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏î‡πâ</p> </div>

<div style="background-color: #faf5ff; border-left: 6px solid #a855f7; padding: 20px; border-radius: 8px;"> <h3 style="color: #4c1d95; margin-top: 0;">üîç 3 ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ Co-training ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à</h3> <div style="margin-bottom: 15px;"> <strong style="color: #6b21a8;">1. Feature Split (‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏õ‡πá‡∏ô‡∏≠‡∏¥‡∏™‡∏£‡∏∞):</strong> <p style="margin: 5px 0; color: #444;">‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡πâ‡∏≠‡∏á‡πÅ‡∏ö‡πà‡∏á‡πÄ‡∏õ‡πá‡∏ô 2 ‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÅ‡∏¢‡∏Å‡∏à‡∏≤‡∏Å‡∏Å‡∏±‡∏ô‡πÑ‡∏î‡πâ (‡πÄ‡∏ä‡πà‡∏ô ‡∏†‡∏≤‡∏û ‡∏Å‡∏±‡∏ö ‡πÄ‡∏™‡∏µ‡∏¢‡∏á) ‡πÇ‡∏î‡∏¢‡∏ó‡∏µ‡πà‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ä‡∏∏‡∏î‡∏ï‡πâ‡∏≠‡∏á "‡πÄ‡∏Å‡πà‡∏á‡∏û‡∏≠" ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏ó‡∏≤‡∏¢‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡πÑ‡∏î‡πâ‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏°‡∏±‡∏ô‡πÄ‡∏≠‡∏á</p> </div> <div style="margin-bottom: 15px;"> <strong style="color: #7e22ce;">2. Peer Review (‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏ó‡∏≤‡∏ô):</strong> <p style="margin: 5px 0; color: #444;">‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 1 ‡∏à‡∏∞‡∏ó‡∏≥‡∏´‡∏ô‡πâ‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡∏Å‡∏≤‡∏£‡∏ö‡πâ‡∏≤‡∏ô‡πÉ‡∏´‡πâ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2 ‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏ó‡∏µ‡πà‡∏°‡∏±‡∏ô‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à ‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2 ‡∏Å‡πá‡∏à‡∏∞‡∏ó‡∏≥‡πÅ‡∏ö‡∏ö‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô‡∏Å‡∏•‡∏±‡∏ö‡∏Ñ‡∏∑‡∏ô‡∏°‡∏≤</p> </div> <div> <strong style="color: #a855f7;">3. Agreement (‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏≠‡∏î‡∏Ñ‡∏•‡πâ‡∏≠‡∏á):</strong> <p style="margin: 5px 0; color: #444;">‡πÄ‡∏õ‡πâ‡∏≤‡∏´‡∏°‡∏≤‡∏¢‡∏Ñ‡∏∑‡∏≠‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏´‡πá‡∏ô‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô‡πÉ‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ Label ‡πÉ‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î</p> </div> </div> </div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.semi_supervised import LabelSpreading
from sklearn.datasets import make_moons

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏π‡∏õ‡∏û‡∏£‡∏∞‡∏à‡∏±‡∏ô‡∏ó‡∏£‡πå‡πÄ‡∏™‡∏µ‡πâ‡∏¢‡∏ß (‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡∏±‡∏ö‡∏ã‡πâ‡∏≠‡∏ô‡∏ó‡∏µ‡πà K-means ‡πÅ‡∏¢‡∏Å‡πÑ‡∏°‡πà‡∏≠‡∏≠‡∏Å)
X, y_true = make_moons(n_samples=200, noise=0.05, random_state=42)

# 2. ‡πÅ‡∏Å‡∏•‡πâ‡∏á‡∏•‡∏ö Label ‡∏≠‡∏≠‡∏Å 90% (‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÅ‡∏Ñ‡πà‡πÑ‡∏°‡πà‡∏Å‡∏µ‡πà‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏õ‡πâ‡∏≤‡∏¢)
y_mixed = np.copy(y_true)
y_mixed[np.random.rand(200) > 0.1] = -1

# 3. ‡πÉ‡∏ä‡πâ Label Spreading "‡∏£‡∏∞‡∏ö‡∏≤‡∏¢‡∏™‡∏µ" ‡∏à‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠
# kernel='knn' ‡∏´‡∏°‡∏≤‡∏¢‡∏ñ‡∏∂‡∏á‡πÉ‡∏´‡πâ‡∏°‡∏≠‡∏á‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏ö‡πâ‡∏≤‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏Å‡∏•‡πâ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î 7 ‡∏Ñ‡∏ô
model = LabelSpreading(kernel='knn', n_neighbors=7)
model.fit(X, y_mixed)

# 4. ‡∏î‡∏∂‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ó‡∏µ‡πà‡∏£‡∏∞‡∏ö‡∏≤‡∏¢‡∏™‡∏µ‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
y_predicted = model.transduction_

# 5. ‡∏ß‡∏≤‡∏î‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_predicted, cmap='viridis', edgecolors='k')
plt.title("Graph-based Label Spreading Result")
plt.show()
```
